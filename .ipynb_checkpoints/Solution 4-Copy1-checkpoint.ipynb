```python
PATTERN = r'[\d\s]+:'
train['Ward_Category'] = train['ward'].str.replace(PATTERN, '', regex = True)
```

```python
def create_code_category(data_frame):
    data_frame['Code_Category'] = data_frame['ADM4_PCODE'].str[:3]
    data_frame['Code_Category'] = data_frame['Code_Category'].map({
        'ZA5' : 'ZA5', 'ZA7' : 'ZA7', 'ZA9' : 'ZA9',
        'ZA6' : 'ZA6', 'ZA8' : 'ZA8', 'ZA4' : 'ZA4',
        'ZA3' : 'ZA3', 'ZA2' : 'ZA2', 'ZA1' : 'ZA1'})
    return data_frame

train = create_code_category(train)
test = create_code_category(test)
```

```python
scaler = StandardScaler(n_features_in_ = 1)
train['ADM4_PCODE'] = scaler.fit_transform(train['ADM4_PCODE'])
```

```python
def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_regression(X, y, discrete_features = discrete_features)
    mi_scores = pd.Series(mi_scores, name = 'MI_scores', index = X.columns)
    mi_scores = mi_scores.sort_values(ascending = False)
    return mi_scores
```

```python
X = train.drop('target', axis = 1)
y = train['target']
discrete_features = X.dtypes == int

mi_scores = make_mi_scores(X, y, discrete_features)
mi_scores
```

```python
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending = True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title('Mutual Information scores')
```

```python
plt.figure(figsize = (20, 15))
plot_mi_scores(mi_scores)
```
